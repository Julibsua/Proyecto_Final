```python
def detectar_outliers_iqr(df):
    print("\n Identificando outliers con el método IQR...\n")
    outliers = {}
    for columna in df.select_dtypes(include=['float64', 'int64']):
        Q1 = df[columna].quantile(0.25)
        Q3 = df[columna].quantile(0.75)
        IQR = Q3 - Q1
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        outliers[columna] = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)][columna].count()
    display(pd.DataFrame(outliers, index=["Cantidad de outliers"]))

print(" Listo para detectar valores atípicos...")
detectar_outliers_iqr(df)
```
